```python
### Taller 1 - Python Básico _ Variables

producto = input("Ingrese el nombre del producto: ")
precio = float(input("Ingrese el precio del producto: "))
cantidad = int(input("Ingrese la cantidad del producto: "))
total = precio * cantidad
print(f"El total a pagar por {cantidad} unidades de {producto} es: {total}")
```

```python
### Estructuras de Datos (Listas, Tuplas, Diccionarios)

articulos_perm = ("Queso", "Leche", "Pan", "Huevos", "Arroz")
Venta1 = {"Producto": "Queso", "Precio": 50.0, "Cantidad": 2, "Unidad de medida": "Kilos"}
Venta2 = {"Producto": "Leche", "Precio": 20.0, "Cantidad": 3, "Unidad de medida": "Litros"}
Venta3 = {"Producto": "Pan", "Precio": 10.0, "Cantidad": 5, "Unidad de medida": "Unidades"}
registro_ventas = [Venta1, Venta2, Venta3]
for venta in registro_ventas:
    print(f"Producto: {venta['Producto']}, Precio: {venta['Precio']}, Cantidad: {venta['Cantidad']}, Unidad de Medida: {venta['Unidad de medida']}")
```

```python
### Estructuras de Control (Condicionales y Ciclos)
registro_ventas = [
    {"Producto": "Mouse", "Precio": 50000, "Cantidad": 2},
    {"Producto": "Teclado", "Precio": 120000, "Cantidad": 1},
    {"Producto": "Tarjeta de Video", "Precio": 300000, "Cantidad": 1}
]
for venta in registro_ventas:
    total_venta = venta["Precio"] * venta["Cantidad"]
    if total_venta > 100000:
        descuento = total_venta * 0.1
        total_final = total_venta - descuento
    else:
        total_final = total_venta
    print("producto:", venta["Producto"])
    print("cantidad:", venta["Cantidad"])
    print("precio unitario:", venta["Precio"])
    print("Subtotal: ", total_venta)
    print("Total con descuento: ", total_final)
    print("-------------")
```

```python
### Funciones y Manejo de Errores
def calcular_ingreso (precio, cantidad):
    try:
        ingreso = float(precio) * float(cantidad)
        return ingreso
    
    except:
        print("Error: Ingrese solo valores numéricos")
        return 0
print(calcular_ingreso(70000, 5))
print(calcular_ingreso("setentamil", 5))

    
```
